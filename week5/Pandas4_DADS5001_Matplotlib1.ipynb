```python
# This is the interface that we will use most often in matplotlib
import matplotlib.pyplot as plt
print(f"Matplotlib backend: {mpl.get_backend()}")
```

Choose figure styles in https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

```python
plt.style.use('seaborn-v0_8')   # 'default', 'classic', 'dark_background', etc.
```

Plotting

```python
# Prepare the data to plot as NumPy's ndarray
x = np.linspace(0,10,100)   # Return evenly spaced numbers over a specified interval
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot y versus x as lines and/or markers
plt.plot( x, y_sin, '-' )
plt.plot( x, y_cos, '--' )

# Display the current figure
# Some Python IDEs (e.g., jupyter) and backends can display plot without calling plt.show()
# BUT some Python IDEs (e.g., idle) and backends (e.g., QtAgg) cannot display plot without calling plt.show()
# Multiple calls to plt.show() in one session may cause backend-dependent unpredictable behaviors
plt.show()
```
