```python
import pytchat
import emoji
import pandas as pd

def clean_message(msg):
    return msg.replace("::", ": :")

chat_data = []

chat = pytchat.create(video_id="8IlNjufJb3Y")
while chat.is_alive():
    for c in chat.get().sync_items():
      message = clean_message(emoji.emojize(c.message, language='alias'))   #‡πÄ‡∏û‡∏¥‡πà‡∏° language='alias'
      print(f"{c.datetime} [{c.author.name}]- {message}") #c. ‡∏≠‡∏≠‡∏Å ‡∏à‡∏≤‡∏Å Messege

      chat_data.append({
          'datetime': c.datetime,
          'author': c.author.name,
          'message': message #c. ‡∏≠‡∏≠‡∏Å
          })

df = pd.DataFrame(chat_data)
df.to_csv('Taengmo_live_chat.csv', index=False, encoding='utf-8-sig')

from google.colab import files
files.download("Taengmo_live_chat.csv")

print("\n‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÑ‡∏ü‡∏•‡πå‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß: youtube_chat.csv")
```
